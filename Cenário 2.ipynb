```python
# Pacote Grover desenvolvido por Filipe Fernandes
# Seu objetivo √© construir o circuito para o algoritmo de Grover de acordo com a string de bits informada
from Grover import GroverCircuit
```

```python
grover = GroverCircuit("0011",True) # True inverte a ordem dos bits para facilitar a leitura
qc = grover.draw()
qc.draw('mpl')
```

```python
# Pacote Relat√≥rio desenvolvido por Filipe Fernandes
# Seu objetivo √© fornecer mecanismos facilitados para execu√ß√£o de circuitos qu√¢nticos, bem como o relat√≥rio de desempenho
from Relatorio import Relatorio
```

```python
# Execu√ß√£o do circuito
dados_execucao = Relatorio.executar_circuito(qc, shots=1024)

# Informa√ß√µes do ambiente
env_info = Relatorio.coletar_info_ambiente(dados_execucao["backend"])

# Gera√ß√£o do relat√≥rio
Relatorio.gerar_relatorio_desempenho(
    num_qubits=qc.num_qubits,
    shots=1024,
    execution_time=dados_execucao["execution_time"],
    env_info=env_info
)

# Resultados do circuito
counts = dados_execucao["result"].get_counts()
```
