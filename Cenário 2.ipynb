```python
# Pacote Grover desenvolvido por Filipe Fernandes
# Seu objetivo é construir o circuito para o algoritmo de Grover de acordo com a string de bits informada
from Grover import GroverCircuit
```

```python
# a partir do oitavo estado, no relatório da célula 21 mostra que estou buscando um estado diferente. REVER ISSO!
alvo = "1111111"
```

```python
grover = GroverCircuit(alvo,True) # True inverte a ordem dos bits para facilitar a leitura
qc = grover.draw()
qc.draw('mpl')
```
